```python
import tensorflow as tf
import matplotlib.pyplot as plt

mnist = tf.keras.datasets.mnist
train_data, test_data = mnist.load_data()
x_train, y_train = train_data
x_test, y_test = test_data
x_train= x_train / 255.0
x_test = x_test / 255.0

model = tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(28, 28)),tf.keras.layers.Dense(128, activation='relu'),
                                    tf.keras.layers.Dense(64, activation='relu'),tf.keras.layers.Dense(10, activation='softmax')])

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
fm = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

plt.plot(fm.history['loss'], label='Training Loss')
plt.plot(fm.history['val_loss'], label='Testing Loss')
plt.xlabel('Number of training iterations')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
fig1, (ax3, ax4) = plt.subplots(1, 2, figsize=(10, 5))
ax1.plot(fm.history['loss'], label='Training Loss')
ax1.set_xlabel('Training iteration')
ax1.set_ylabel('Training Loss')
ax1.legend()

ax2.plot(fm.history['val_loss'], label='Testing Loss')
ax2.set_xlabel('Training iteration')
ax2.set_ylabel('Testing Loss')
ax2.legend()

ax3.plot(fm.history['accuracy'], label='Training Accuracy')
ax3.set_xlabel('Training iteration')
ax3.set_ylabel('Training accuracy')
ax3.legend()

ax4.plot(fm.history['val_accuracy'], label='Testing Accuracy')
ax4.set_xlabel('Testing iteration')
ax4.set_ylabel('Testing accuracy')
ax4.legend()

plt.show()
```
